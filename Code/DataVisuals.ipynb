```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
QB_file = pd.read_csv("/content/finalQB.csv")
QB_file
```

```python
QB_file.plot(x='Sk%', y='%Cap', kind='scatter')
plt.xlabel('Percentage of Times Sacked (Z-Score)')
plt.ylabel('Second Contract Salary (Millions)')
plt.title('QB Sack Percentage by Second Contract Salary')
plt.show()
```

```python
QB_file.plot(x='TD', y='%Cap', kind='scatter')
plt.xlabel('Touchdowns by Quarterbacks (Z-Score)')
plt.ylabel('Second Contract Salary (Millions)')
plt.title('QB Passing Touchdowns by Second Contract Salary')
plt.show()
```

```python
QB_file.plot(x='Rk', y='%Cap', kind='scatter')
plt.xlabel('Quarterback Season Rank (Z-Score)')
plt.ylabel('Second Contract Salary (Millions)')
plt.title('QB Season Rank by Second Contract Salary')
plt.show()
```

```python
QB_file.Con_Sal.plot.density(color='green')
plt.title('Density plot for Quarterback Second Contract Salary')
plt.show()
```
